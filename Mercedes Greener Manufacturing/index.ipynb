# Dimensionality Reduction

```python
train_=train
test_=test
x_train=train.iloc[1:,:].values
x_test=test.iloc[:,:].values
```

```python
from sklearn.preprocessing import scale
```

```python
x_train_std=scale(x_train)
x_test_std=scale(x_test)
```

```python
from sklearn.decomposition import PCA
```

```python
pca_train=PCA()
pca_train.fit(x_train_std)

var_train=np.cumsum(np.round(pca_train.explained_variance_ratio_,decimals=4)*100)

plt.plot(var_train)
plt.axhline(y=80,ls=':',color='black')
plt.axvline(x=135,ls=':',color='black')
plt.title('Scree Plot for Train')
plt.xlabel('Number of Components')
plt.ylabel('Percentage of Variation Explained')
```

```python
pca_test=PCA()
pca_test.fit(x_test_std)

var_test=np.cumsum(np.round(pca_test.explained_variance_ratio_,decimals=4)*100)

plt.plot(var_test)
plt.axhline(y=80,ls=':',color='black')
plt.axvline(x=135,ls=':',color='black')
plt.title('Scree Plot for Test')
plt.xlabel('Number of Components')
plt.ylabel('Percentage of Variation Explained')
```

```python
pca_train=PCA(n_components=135)
pca_train.fit(x_train_std)

x_train_final=pca_train.transform(x_train_std)
```

```python
pca_test=PCA(n_components=135)
pca_test.fit(x_test_std)

x_test_final=pca_test.transform(x_test_std)
```
